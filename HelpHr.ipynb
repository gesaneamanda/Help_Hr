```python
!pip install nltk

import nltk
from nltk.stem import RSLPStemmer

nltk.download('rslp')
nltk.download('punkt')

# --- Função para processar a pergunta do usuário ---
def processar_pergunta(pergunta):
  """
  Processa a pergunta do usuário:
    1. Tokeniza a pergunta em palavras.
    2. Reduz as palavras ao seu radical (stemming).
    3. Retorna a lista de radicais.
  """
  tokens = nltk.word_tokenize(pergunta.lower())
  stemmer = RSLPStemmer()
  radicais = [stemmer.stem(token) for token in tokens]
  return radicais

# --- Exemplo de uso ---
pergunta = "Qual o meu saldo de férias?"
radicais = processar_pergunta(pergunta)
print(radicais)  # Output: ['qual', 'o', 'meu', 'sald', 'de', 'feri']

# --- Próximos passos ---
# 1. Criar uma base de conhecimento com perguntas e respostas.
# 2. Implementar lógica para encontrar a resposta mais relevante
#    com base nos radicais da pergunta.
# 3. Integrar com uma plataforma de chatbot (Dialogflow, etc.).
```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]

  produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Conteudo"]
```

```python
consulta = "Quais documentos que devo levar na empresa?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```

```python
prompt = f"Reescreva esse texto de uma forma mais descontraída, com emojis, para que a pessoa se sinta acolhida {trecho}"

model_2 = genai.GenerativeModel("gemini-1.0-pro",
                                generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
